## Question 1

Following are the five plots that we can plot using the Seaborn Library: - 
1. **Scatterplot:** A scatterplot is used to visualize the relationship between two continuous variables. Each point on the plot represents an observation in the dataset.

2. **Relplot (Relational plot):** Relplot is a versatile function that can create different types of plots (e.g., scatter plots, line plots, etc.) to show the relationship between two variables across multiple subsets of the data.

3. **Catplot (Categorical plot):** Catplot is used for plotting the relationship between one categorical variable and another variable (either categorical or numerical). It can create different types of plots like box plots, violin plots, etc.

4. **Pairplot:** Pairplot is used for plotting pairwise relationships in a dataset. It creates a matrix of scatterplots where each variable is compared with every other variable.

5. **Jointplot:** Jointplot is used to visualize the joint distribution of two variables along with the univariate distributions of each variable. It combines scatter plots, histograms, and kernel density plots.
