## Question 1 

Estimation statistics is a branch of statistics that dels with the process of estimating unknown parameters of a population based on sample data. 

There are two main types of estimation in statistics:
1. Point estimate: - It is a single value that serves as the best guess or estimate of an unknown population parameter. 

2. Interval Estimate: - It provides a range of values within which the true value of the population parameter is likely to lie, along with a level of confidence associated with that range. 

## Question 2

```python
import numpy as np 
```

```python
def population_estimation(sample_mean,sample_std,sample_size):
    standard_error = sample_std/np.sqrt(sample_size)
    margin_of_error = 1.96 * standard_error
    lower_bound = sample_mean-margin_of_error
    upper_bound = sample_mean+margin_of_error
    return (lower_bound,upper_bound)
```

```python
LowerBound,UpperBound = population_estimation(50,50,100)
print(f"Estimated population mean of 95% confidence interval is: {LowerBound} - {UpperBound}")
```

## Question 3

Hypothesis testing is a statistical method used to make inferences about population parameters based on sample data.

The basic steps of hypothesis testing typically include:

- Formulating a null hypothesis (H0) and an alternative hypothesis (Ha).
- Selecting an appropriate statistical test and significance level (alpha, denoted as α).
- Collecting sample data and calculating a test statistic based on the sample.
- Comparing the test statistic to a critical value or p-value to determine whether to reject the null hypothesis.

Hypothesis Testing is used for various purposes in research, science, and decision making.

```python
import numpy as np 
from scipy import stats
```

```python
def independent_ttest(sample1,sample2,alpha=0.05):
    t_statistics,p_value = stats.ttest_ind(sample1,sample2)
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Accept the null hypothesis")
```

## Question 11

Student's t-distribution, often referred to simply as the t-distribution, is a probability distribution that arises from the estimation of the mean of a normally distributed population when the sample size is small and the population standard deviation is unknown.

Here are some key points about the t-distribution and when to use it:

Small Sample Sizes: The t-distribution is particularly useful when dealing with small sample sizes (typically n < 30) where the population standard deviation is unknown. In such cases, using the sample standard deviation to estimate the population standard deviation leads to a slightly different distribution of sample means.

Population Standard Deviation Unknown: Unlike the normal distribution, which assumes a known population standard deviation, the t-distribution does not require knowledge of the population standard deviation. It is robust against the uncertainty in estimating the population standard deviation from the sample.

Confidence Intervals and Hypothesis Testing: The t-distribution is commonly used in constructing confidence intervals and conducting hypothesis tests for the population mean when the sample size is small and/or the population standard deviation is unknown. It provides a more accurate estimate of the sampling distribution of the sample mean in such scenarios.

Comparison of Means: When comparing means of two independent samples or conducting a one-sample t-test, the t-distribution is used to calculate the test statistic and assess the significance of the difference between means.

Regression Analysis: In linear regression analysis, the t-distribution is used in hypothesis testing for the regression coefficients when the sample size is small.

## Question 12 

The t-statistic is a measure used in hypothesis testing to determine if there is a significant difference between the means of two groups or if a sample mean differs significantly from a population mean. It is essentially a standardized measure of the difference between sample means, taking into account the variability within the samples.

The formula for calculating the t-statistic depends on the context in which it is used. Here are the formulas for two common scenarios:

1. **One-Sample t-Test**: When comparing the mean of a single sample to a known or hypothesized population mean, the formula for the t-statistic is:

   \[ t = \frac{{\bar{x} - \mu}}{{s / \sqrt{n}}} \]

   Where:
   - \( \bar{x} \) is the sample mean,
   - \( \mu \) is the population mean (or the hypothesized population mean under the null hypothesis),
   - \( s \) is the sample standard deviation,
   - \( n \) is the sample size.

2. **Two-Sample t-Test**: When comparing the means of two independent samples, the formula for the t-statistic is:

   \[ t = \frac{{\bar{x}_1 - \bar{x}_2}}{{s_p \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}} \]

   Where:
   - \( \bar{x}_1 \) and \( \bar{x}_2 \) are the sample means of the two samples,
   - \( s_p \) is the pooled standard deviation (a measure of the combined variability of the two samples),
   - \( n_1 \) and \( n_2 \) are the sample sizes of the two samples.

## Question 14 

```python
import numpy as np 
from scipy.stats import t 

#Given Data 
sample_mean = 8
hypothesis_mean = 10
sample_std = 3
sample_size = 100
significance_level = 0.05

t_statistics = (sample_mean-hypothesis_mean)/(sample_std/np.sqrt(sample_size))
degree_of_freedom = sample_size-1
critical_value = t.ppf(significance_level,degree_of_freedom)
if t_statistics < critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

```python
import numpy as np
from scipy.stats import t

# Given data
mean1 = 80
std1 = 10
n1 = 30

mean2 = 75
std2 = 8
n2 = 40

significance_level = 0.01


sp = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))


t_statistic = (mean1 - mean2) / (sp * np.sqrt(1/n1 + 1/n2))


degrees_of_freedom = n1 + n2 - 2


critical_value = t.ppf(1 - significance_level/2, degrees_of_freedom)


if abs(t_statistic) > critical_value:
    print("Reject the null hypothesis. The population means for the two groups are not equal.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest that the population means for the two groups are not equal.")
```
