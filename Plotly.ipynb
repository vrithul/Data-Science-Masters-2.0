```python
import seaborn as sns 
import plotly.express as px 
```

## Question 1 

```python
titanic = sns.load_dataset('titanic')
titanic 
```

```python
fig = px.scatter(titanic,x='age',y='fare')
fig.update_layout(title="Scatter Diagram of Fare among Age",xaxis_title="Age",yaxis_title="Fare")
fig.show()
```

## Question 2 

```python
tips = px.data.tips()

tips
```

```python
fig = px.box(tips,x='day',y='total_bill')
fig.update_layout(title="Box Plot of Total Bill Amount per Day",xaxis_title="Day",yaxis_title="Total Bill Amount")
fig.show()
```

## Question 3 

```python
fig = px.histogram(tips,x='sex',y='total_bill',color='day',pattern_shape='smoker')
fig.update_layout(title="Histogram ")
fig.show()
```

## Question 4 

```python
iris = px.data.iris()
iris
```

```python
fig = px.scatter_matrix(iris,dimensions=['sepal_length','sepal_width','petal_length','petal_width'],color='species')
fig.update_yaxes(automargin=True)
fig.update_layout(title="Scatter Matrix",height=600)
fig.show()
```

## Question 5 


Displot (also known as a distribution plot) is used to represent data in histogram form. It is a univariant set of collected data, which means the data distribution of one variable will be shown against another variable. In Python, we use the Seaborn library with Matplotlib for data visualization.

```python
fig = px.histogram(tips, x='total_bill', marginal='rug', nbins=30)
fig.update_layout(title="Distribution Plot")
fig.show()
```
