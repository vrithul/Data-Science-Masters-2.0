**Question 7**

```python
df = pd.DataFrame({'Email':['v.rithul@mca.christuniversity.in','yash@mca.christuniversity.in','vrithul9@gmail.com','radhisharajesh@gmail.com']})
print(df)

import re
pattern = r'^([^@]+)'
username = []
def usernames(data):
    for i in df['Email']:
        a = re.search(pattern,i)
        username.append(a.group(1))
    df['Username'] = username

usernames(df)
print('\n')
print(df)

```

**Question 10**

```python

data = {'date': pd.date_range(start='2023-10-01', periods=10, freq='D'),
        'sales': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]}
df = pd.DataFrame(data)


def calculate_moving_average(df, window_size=7):
    df['Moving average'] = df['sales'].rolling(window=window_size, min_periods=1).mean()
    return df


df = calculate_moving_average(df)


print(df)
```

**Question 12**

```python

data = {'date': pd.date_range(start='2023-01-01', periods=10, freq='D')}
df = pd.DataFrame(data)

def select_rows_in_date_range(df, start_date, end_date):
    mask = (df['date'] >= start_date) & (df['date'] <= end_date)
    selected_rows = df[mask]
    return selected_rows

start_date = '2023-01-02'
end_date = '2023-01-31'

selected_data = select_rows_in_date_range(df, start_date, end_date)

print(selected_data)
```
